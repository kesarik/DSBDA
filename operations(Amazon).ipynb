```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Amazonbookreviews.csv")
df
```

```python
#datasubset for row wise
df1=df.iloc[:20,:]
df1
```

```python
df2=df.iloc[21:21,:]
df2
```

```python
#row merged
merged=pd.concat([df1,df2],axis=0)
merged
```

```python
#subset using columnwise
df3=df.loc[:,['reviewerID','reviewername']]
df3
```

```python
df4=df.loc[:,['helpful','reviewerText']]
df4
```

```python
merged_column=pd.concat([df3,df4])
merged_column
```
