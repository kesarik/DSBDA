```python
df=pd.read_excel("AirQualityUCI.xlsx")
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
plt.title("Box Plot")
sb.boxplot(x="PT08.S3(NOx)",data=df)
```

```python
plt.title=("Histogram")
sb.histplot(x="NOx(GT)",data=df,bins=40)
plt.xlabel("NOx(GT)")
plt.ylabel("Frequency")
plt.show()
```

```python
Date_data=df[df['Date']=="2004-03-11"]
plt.title("Single Line Graph")
plt.plot(Date_data['CO(GT)'])
plt.xlabel("Index Of Date")
plt.ylabel("CO(GT)")
plt.show()
```

```python
import seaborn as sns
```

```python
Date_data = df[df['Date'].isin(["2004-03-11", "2004-03-15"])]
plt.title("Multiline Graph")
sns.lineplot(data=Date_data, x="CO(GT)", y="PT08.S1(CO)", hue="Date")
plt.show()
```
